# Delayed Information Distributed by Tail Number Monthly

Tail number is donated to a registered ID of a aircraft.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
```

```python
%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
airplanesRecord = pd.read_csv('assets/data/DelayedFlights.csv')
#airplanesRecord.head()
#airplanesRecord.info()
```

```python
groupedByYear = airplanesRecord.groupby(by='Year')
```

```python
groupedByYearCount = groupedByYear.count()
#groupedByYearCount
```

```python
groupedByMonth = airplanesRecord.groupby(by='Month').count()
#groupedByMonth
```

```python
tailNumberMonth = airplanesRecord.groupby(by=['TailNum','Month']).count()['FlightNum'].unstack()
```

```python
plt.figure(figsize=(25,20))
plt.tight_layout()
sns.heatmap(tailNumberMonth)
```
